```python
import numpy as np
import re
import pandas as pd 
from pyknp import KNP

#テキストの分割

slice_num = 4

f = open('bad/seven.txt', encoding='utf-8')
lines = f.read()
f.close()

#lines = [i for i in lines if not i == '\n']
#lines = re.findall(r'[^。]+(?:[。]|$)', lines)

lines = re.split('[。,」]', lines)

sn = round(len(lines)/slice_num)
lines = lines[-sn:-1]
#lines = lines[-5:-3]
```

```python
character = "ぼく"

#f = open('test1.txt', encoding='utf-8')
#lines = f.readlines()
#f.close()


#text = text.replace('\u3000', '')

knp = KNP()

def extract_repname(text: str) :
    tag_list = knp.parse(text).tag_list()
    tag_repname = list()
    for tag in tag_list:  # 各基本句へのアクセス
        if re.search('<正規化代表表記', tag.fstring):
            if re.search("ガ:"+character, tag.fstring):
                tag_repname.append(tag.normalized_repname)
    return tag_repname

repname_list = []
for line in lines:
    #line = line.replace('\n', '')
    repname_list.append(extract_repname(line))

res = [ flatten for inner in repname_list for flatten in inner ]
a = pd.DataFrame(res)
splited = a[0].str.split('/', expand=True)
splited = splited.rename(columns={0:'word'})
splited
```

```python
features = ['word', 'yomi', 'p', 'n', 'cate1', 'cate2']
posi_nega_dic = pd.read_csv("posi_nega_dic.csv", names=features)

morpho_analysed_all = pd.merge(splited, posi_nega_dic, on = ['word'], how = 'left')
morpho_analysed_all = morpho_analysed_all.reset_index()

analysed = morpho_analysed_all.dropna(subset=['p'])
analysed
```
